```
from selenium import webdriver
import openpyxl
import time
import re

try:
    wb = openpyxl.load_workbook("LGwebsite.xlsx")
    sheet = wb.active
    print("불러오기 완료")
except:
    wb = openpyxl.Workbook()
    sheet = wb.active
    sheet.append(["question", "user", "category", "detail1", "detail2"])
    print("새로운 파일을 만들었습니다.")

driver = webdriver.Chrome("../chromedriver 6")
driver.get("http://www.uplus.co.kr/css/orub/erms/FaqList.hpi?gubun=0&mid=10749")
time.sleep(3)
id = 0
page = 1

while True:

    print('page:',page)
    buttons = driver.find_elements_by_css_selector('span.number a')
    container = driver.find_elements_by_css_selector('ul.js_accordion li')

    for cont in container:
        data = cont.find_element_by_css_selector('div.title>a').text
        data = re.sub('(\[([^>]+)\])', '', data).strip()
        cut_point = data.find('개인')

        question = data[:cut_point].strip()
        category = data[cut_point:].strip()

        lst = category.split(' > ')
        lst.insert(0, question)
        sheet.append(lst)
        #print(lst)

    try:
        if id%9==0 and id!=0:
            next_button = driver.find_element_by_css_selector('span.next a')
            next_button.click()
            time.sleep(3)
            id = 0
        else:
            buttons[id].click()
            time.sleep(3)
            id += 1
    except:
        break

    page += 1

driver.close()
wb.save("LGwebsite.xlsx")
```
